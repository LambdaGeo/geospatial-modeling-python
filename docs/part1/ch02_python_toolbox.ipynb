## Data as Coordinates

In the world of computing, data refers to any collection of values or information that can be processed, stored, and transformed by a computer. This data can represent a wide variety of information, including text, images, and coordinates. In reality, any information processed by a computer must eventually be translated into numbers, as computers remain, at their core, sophisticated machines for numerical processing. In geography, everything starts with a location. To a computer, a location is simply a set of numbers stored in variables. A variable acts as a labeled container for data.
